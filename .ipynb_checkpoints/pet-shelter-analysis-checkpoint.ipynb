# Is there an area where more pets are found?

```python
locations = intakes_and_outcomes['found_location'].nunique() # to find the unique location in the dataset
locations
```

```python
animals_location = intakes_and_outcomes.groupby('found_location').agg({'animal_type': 'count'})
animals_location.columns = ['num_animal_types']
animals_location = animals_location.reset_index()
animals_location.sort_values(by='num_animal_types', ascending=False)[:5]
```

```python
sns.barplot(data=locations, x='num_animal_types', y='found_location',)
plot.show()
```
