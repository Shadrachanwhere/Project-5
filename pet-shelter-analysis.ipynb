# What is the average number of pets found in a month in the year 2015?
The average number of pets found in a month in the year 2015 is 1558.
First, I filtered the dataset per intake year which is 2015. i counted the values per intake month and use the mean function
# Are there months where there is a higher number of animals found? 
Yes.
There is a higher  number of animals found within the following months; May, June, July, August, September and October. These months have a high number of found animals higher than the average which is 1558.
I achieved this by comparing the number of found animals in each month with the average.


```python
pets_2015 = intakes_and_outcomes[intakes_and_outcomes['intake_year'] == 2015]
monthly_totals = pets_2015['intake_month'].value_counts()
average_intake = monthly_totals.mean()
average_intake

above_average_months = monthly_totals[monthly_totals.values > average_intake]
above_average_months.sort_index()
```

# What is the ratio of incoming pets vs. adopted pets?
The ratio of incoming pets vs adopted pets is 2. Meaning for a every 2 pets that enter the shelter only one pet is adopted.
i did this by counting the number of adopted pets and intake pets and dividing the number of incoming pets by the number of adopted pets(number of incoming_pets / number of adopted_pets)


```python
adopted_pets = intakes_and_outcomes[intakes_and_outcomes['outcome_type'] =='Adoption']
num_adopted_pets = len(adopted_pets)
incoming_pets = len(intakes)
ratio = incoming_pets / num_adopted_pets
round(ratio)
```

# What is the distribution of the types of animals in the shelter?
As we can see below the animal type and their respective number. The most common animal found in the shelter is Dog.
I achieved this by grouping all the animal types in the shelter and aggregating(count) according to their type or kind.

```python
animal_distribution = intakes_and_outcomes.groupby('animal_type').agg({'animal_id_intake': 'count'})
animal_distribution.columns = ['Number of types of animals in the shelter']
animal_distribution  = animal_distribution.reset_index()
animal_distribution
```

# What are the adoption rates for specific breeds?
The top 5 Breed of Dogs are Labrador Retriever Mix, Pit Bull Mix,Chihuahua Shorthair Mix, 
German Shepherd mix and Austrilain cattle Dog Mix. Among the 5 top breed of dogs, Ladbrador Retriever Mix being the most adopted breed with  **23,688** number of adoption, with a percentage of **27.9%**.
I achieved this by first filtering dataset, the columns; animal_type and the outcome_type to be dog and adoption respectively.
After finding dogs who are adopted, i further grouped them per breed and aggregated them using the count function.
I then sorted the values in respect to their number of adoption.
Secondly, to find the ratio, I created another column by dividing (the number of adoption / the sum of adoption) and multiplying 
it by 100 to find percentage.

```python
dogs = intakes_and_outcomes[intakes_and_outcomes['animal_type'] == 'Dog']
adoption = dogs[dogs['outcome_type'] == 'Adoption']
breed = adoption.groupby('breed').agg({'animal_id_intake': 'count'})
breed.columns = ['number of adoption']
breed = breed.reset_index()
top_breeds = breed.sort_values(by='number of adoption', ascending=False)
top_breeds['percentage'] = round((top_breeds['number of adoption'] / top_breeds['number of adoption'].sum()) * 100,1)

top_breeds[:5]
```

# What are the adoption rate for different colorings?
The top 5 adoption rate for different colorings are below.
black/white being the most favourite color of adoption number **3,738** and adoption rate of  **11.1%**.
And brown/white being the least among the 5 colorings. I achieved this by filtering the column(outcome_type) of the dataset to adoption and grouping it according to their color and using the aggregate(count) function on them.
I then created another column to find the percentage by (number of colors / the sum of number of colors) and multiplying the
results by 100.

```python
animals_adopted = intakes_and_outcomes[intakes_and_outcomes['outcome_type'] == 'Adoption']
coloring = animals_adopted.groupby('color').agg({'animal_id_intake': 'count'})
coloring.columns = ['num of colors']
coloring = coloring.reset_index()
top_coloring = coloring.sort_values(by='num of colors', ascending=False)
top_coloring['percentage'] = round((top_coloring['num of colors'] / top_coloring['num of colors'].sum()) * 100,1)
top_5 = top_coloring[:5]
top_5
```

# About how many animals are spayed/neutered each month?
About 4,085 animals are spayed/neutered each month. I did this by filtering the column(sex_upon_intake) to intact male or intact female,
after, I grouped the newly filtered dataset into intake month and counted the animal_id_intake column.

```python
male_female_intact = intakes_and_outcomes[(intakes_and_outcomes['sex_upon_intake'] == 'Intact Female') | (intakes_and_outcomes['sex_upon_intake']=='Intact Male')]
each_month = male_female_intact.groupby('intake_month').agg({'animal_id_intake': 'count'})
average_per_month = round(each_month['animal_id_intake'].mean())
average_per_month
```
